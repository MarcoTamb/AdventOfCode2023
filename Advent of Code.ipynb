```python
from itertools import product

def check_number_left(row, column):
    if data_raw[row][column] not in '0123456789':
        return 0
    try:
        not_last_digit=data_raw[row][column+1] in '0123456789'
    except IndexError:
        not_last_digit=False
        pass
    if not_last_digit:
        return 0
    number=re.findall('\d*$', data_raw[row][:column+1])[0]
    return int(number)
def check_number_right(row, column):
    if data_raw[row][column] not in '0123456789':
        return 0
    try:
        not_first_digit=data_raw[row][column-1] in '0123456789'
    except IndexError:
        not_first_digit=False
        pass
    if not_first_digit:
        return 0
    number=re.findall('^\d*', data_raw[row][column:])[0]
    return int(number)

class gear():
    def __init__(self, row, column):
        self.row=row
        self.column=column
        self.first_number=0
        self.second_number=0
        self.too_many_numbers=False
    def found_number(self, number):
        if number!=0:
            print(number)
            if self.first_number==0:
                self.first_number=number
                return
            elif self.second_number==0:
                self.second_number=number
                return
            else:
                self.too_many_numbers=True
    def check_adjacent(self):
        for row, column in product([self.row-1,self.row, self.row+1],[self.column-1,self.column, self.column+1]):
            number=check_number_left(row, column)
            if number:
                self.found_number(number)
        for row, column in product([self.row-1,self.row, self.row+1],[self.column-1,self.column, self.column+1]):
            number=check_number_right(row, column)
            if number not in [self.first_number, self.second_number]: # a bit greedy: assuming the two numbers in the pairs are distinct, but I don't want to implement the logic to check the start and end of the number
                self.found_number(number)
    def product(self):
        product=self.first_number * self.second_number * int(not self.too_many_numbers)
        print(f'{self.first_number}, {self.second_number}, {self.too_many_numbers}, {product}')
        return product

list_of_products=[]
for asterisk in asterisks:
    print(asterisk)
    current_gear=gear(asterisk[0], asterisk[1])
    current_gear.check_adjacent()
    list_of_products.append(current_gear.product())
sum(list_of_products)
```

```python
def convert_tuples(map, list_of_tuples):
    if list_of_tuples==[]:
        return []
    converted=None
    line_conversion=None
    converting_tuple=list_of_tuples[0]
    for line in maps[map]:
        if converting_tuple[0] >= line[1] and converting_tuple[0] < line[1]+line[2]:
            converted=(converting_tuple[0]-line[1])+line[0]
            line_conversion=line
    if converted==None:
        converted=converting_tuple[0]
        differences=[line[1]-converting_tuple[0] for line in maps[map] if line[1]>converting_tuple[0]]
        difference=min(differences) if differences!=[] else max([line[1]+line[2] for line in maps[map]])+converting_tuple[1]
        line_conversion=(converted, converted, difference)
    if converting_tuple[0]+converting_tuple[1]>line_conversion[1]+line_conversion[2]:
        converted_tuple=(converted, line_conversion[1]+line_conversion[2]-converting_tuple[0])
        new_list_of_tuples=[(line_conversion[1]+line_conversion[2], converting_tuple[1]-converted_tuple[1])]+list_of_tuples[1:]
    else:
        converted_tuple=(converted, converting_tuple[1])
        new_list_of_tuples=list_of_tuples[1:]
    return [converted_tuple]+convert_tuples(map, new_list_of_tuples)
```

```python
combinations={}
for seed in seeds:
    combinations[seed]={'seed': [seed]}
    combinations[seed]['soil']=convert_tuples('seed-to-soil',combinations[seed]['seed'])
    combinations[seed]['fertilizer']=convert_tuples('soil-to-fertilizer', combinations[seed]['soil'])
    combinations[seed]['water']=convert_tuples('fertilizer-to-water', combinations[seed]['fertilizer'])
    combinations[seed]['light']=convert_tuples('water-to-light', combinations[seed]['water'])
    combinations[seed]['temperature']=convert_tuples('light-to-temperature', combinations[seed]['light'])
    combinations[seed]['humidity']=convert_tuples('temperature-to-humidity', combinations[seed]['temperature'])
    combinations[seed]['location']=convert_tuples('humidity-to-location', combinations[seed]['humidity'])
```

```python
import numpy as np
import math
with open('input_day7.txt', 'r') as file:
    data_raw = file.read().split('\n')
# data='''32T3K 765
# T55J5 684
# KK677 28
# KTJJT 220
# QQQJA 483
# '''.split('\n')
value_dict = {'T': 10, 'J': 1, 'Q': 12, 'K': 13, 'A': 14}


def convert_number(n):
    if n in '23456789':
        return int(n)
    else:
        return value_dict[n]


def replace_jacks(hand_list, jack, replacing):
    return sum([[hand[:jack]+[replaced]+hand[jack+1:] for hand in hand_list] for replaced in list(replacing)], [])


def calc_hand_value(cards):
    counts = [cards.count(card) for card in cards]
    if 5 in counts:
        hand_value = 6
    elif 4 in counts:
        hand_value = 5
    elif 3 in counts:
        if 2 in counts:
            hand_value = 4
        else:
            hand_value = 3
    elif 2 in counts:
        if counts.count(2) == 4:
            hand_value = 2
        else:
            hand_value = 1
    else:
        hand_value = 0
    return hand_value


class hand():
    def __init__(self, hand, bid):
        self.cards = hand
        self.bid = bid
        self.get_hand_list()
        self.sort_value = self.get_sorting_value()
        self.rank = 0

    def get_hand_list(self):
        cardset = set(self.cards)
        if 1 not in cardset:
            hand_list = [self.cards]
        else:
            hand_list = [self.cards]
            jacks = [i for i, x in enumerate(self.cards) if x == 1]
            for jack in jacks:
                hand_list = replace_jacks(
                    hand_list, jack, [card for card in cardset if card != 1]+[14])
        self.hand_list = hand_list

    def get_winnings(self):
        return self.rank * self.bid

    def get_sorting_value(self):
        return max([calc_hand_value(cards)*10000000 + sum([self.cards[i]*(15**(4-i)) for i in range(5)]) for cards in self.hand_list])


hands = [hand([convert_number(char) for char in line[:5]], int(line[6:]))
         for line in data_raw if line != '']

hands = sorted(hands, key=lambda a: a.sort_value)
for index, hand in enumerate(hands):
    hand.rank = index+1
print(sum([hand.get_winnings() for hand in hands]))
```

### Problem 1

```python
def find_reflection(pattern):
    rows, cols=len(pattern), len(pattern[0])
    for i in range(rows-1):
        if i+1<rows/2:
            if all([element1==element2 for element1, element2 in zip(pattern[:i+1], pattern[i+1:2*i+2][::-1])]):
                return (i+1, 'V')
        else:
            if all([element1==element2 for element1, element2 in zip(pattern[2*i-rows+2:i+1],  pattern[i+1:][::-1])]):
                return (i+1, 'V')
    for i in range(cols-1):
        if i+1<cols/2:
            zipped=zip([line[:i+1] for line in pattern], [line[i+1:2*i+2][::-1] for line in pattern])
        else:
            zipped=zip([line[2*i-cols+2:i+1] for line in pattern], [line[i+1:][::-1] for line in pattern])
        if all([element1==element2 for element1, element2 in zipped]):
            return (i+1, 'H')

    return 'error, no line found'
```

```python
reflections=[find_reflection(pattern) for pattern in patterns]
vertical=[reflection[0] for reflection in reflections if reflection[1]=='V']
horizontal=[reflection[0] for reflection in reflections if reflection[1]=='H']
sum(horizontal)+100*sum(vertical)
```
