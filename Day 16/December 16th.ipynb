```python
import itertools
with open('input_day16.txt', 'r') as file:
    data=file.read()

# data='''.|...\....
# |.-.\.....
# .....|-...
# ........|.
# ..........
# .........\\
# ..../.\\\\..
# .-.-/..|..
# .|....-|.\\
# ..//.|....
# '''
data=data.split('\n') 
data=['#'*(len(data[0])+2)]+['#'+line+'#' for line in data if line !='']+['#'*(len(data[0])+2)]
def is_not_edge(t):
    if ((0 not in t) and (len(data)-1!=t[0]) and (len(data[0])-1!=t[1])): 
        return True
    else: 
        return False

for line in data:
    print(line)
```

### Problem 1

```python
def add_to_coord(line, col, direction):
    encounter=data[line][col]
    if (line, col, direction) in visited:
        loop.append((line, col, direction))
        return []
    else:
        visited.append((line, col, direction))
    if encounter=='.':
        return [(line, col, direction)]
    elif encounter=='#':
        return []
    elif encounter=='|':
        if direction in 'LR':
            return [(line, col, 'U'), (line, col, 'D')]
        else: 
            return [(line, col, direction)]
    elif encounter=='-':
        if direction in 'UD':
            return [(line, col, 'L'), (line, col, 'R')]
        else: 
            return [(line, col, direction)]
    elif encounter=='\\':
        if direction in 'U':
            return [(line, col, 'L')]
        elif direction in 'D':
            return [(line, col, 'R')]
        elif direction in 'L':
            return [(line, col, 'U')]
        elif direction in 'R':
            return [(line, col, 'D')]
    elif encounter=='/':
        if direction in 'U':
            return [(line, col, 'R')]
        elif direction in 'D':
            return [(line, col, 'L')]
        elif direction in 'L':
            return [(line, col, 'D')]
        elif direction in 'R':
            return [(line, col, 'U')]
def progress_beam(beam):
    new_beam=[]
    for ray in beam:
        if ray[2]=='R':
            new_coord=(ray[0],ray[1]+1)
            if is_not_edge(new_coord):
                energized[new_coord]+=1
                new_beam=new_beam+add_to_coord(new_coord[0], new_coord[1], 'R')
        if ray[2]=='L':
            new_coord=(ray[0],ray[1]-1)
            if is_not_edge(new_coord):
                energized[new_coord]+=1
                new_beam=new_beam+add_to_coord(new_coord[0], new_coord[1], 'L')
        if ray[2]=='U':
            new_coord=(ray[0]-1,ray[1])
            if is_not_edge(new_coord):
                energized[new_coord]+=1
                new_beam=new_beam+add_to_coord(new_coord[0], new_coord[1], 'U')
        if ray[2]=='D':
            new_coord=(ray[0]+1,ray[1])
            if is_not_edge(new_coord):
                energized[new_coord]+=1
                new_beam=new_beam+add_to_coord(new_coord[0], new_coord[1], 'D')
    return new_beam
```

```python
visited=[]
loop=[]
energized={
    tup: 0
    for tup in itertools.product(range(len(data)), range(len(data[0]))) if is_not_edge(tup)
}
energized[(1, 1)]=1
light_beam=[(1, 0, 'R')]
while light_beam!=[]:
    light_beam=progress_beam(light_beam)
print(len([value for value in energized.values() if value !=0]))
```
